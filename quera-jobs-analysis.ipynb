```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("job_posts.csv")
df
```

```python
df.info()
```

```python
df.dropna(subset=['Salary', 'Level', 'Collaboration Type'], inplace=True)
```

```python
df.info()
```

```python
df['Salary'] = df['Salary'].str.extract(r"(\d+)", expand=False).astype(int)
df['Salary']
```

```python
df['Salary'] = df.groupby('Level')['Salary'].transform(lambda x: x.fillna(x.mean()))
```

```python
start_of_corona = datetime.strptime("2020-02-19", "%Y-%m-%d")
start_of_corona
```

```python
df['Publish Time'] = pd.to_datetime(df['Publish Time'])
df['start_of_corona'] = df['Publish Time'].apply(lambda X: "before" if X<pd.to_datetime('2020-2-19') else "after")
```

```python
df
```

```python
df['remote_flag'] = df['Offers Remote'].apply(lambda x: 1 if x == True else 0)
```

```python
df
```

```python
remote_before = df[(df['start_of_corona'] == 'before' ) & (df['Offers Remote'] == True)].shape[0]
total_before = df[df['start_of_corona'] == 'before'].shape[0]
rate_before = remote_before / total_before
print(rate_before)
```

```python
remote_after = df[(df['start_of_corona'] == 'after' ) & (df['Offers Remote'] == True)].shape[0]
total_after = df[df['start_of_corona'] == 'after'].shape[0]
rate_after = remote_after / total_after
print(rate_after)
```

```python
remote_growth = (rate_after - rate_before)
print(remote_growth)
```
